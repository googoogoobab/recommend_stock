```python
from pandas_datareader import data as pdr
import yfinance as yf
yf.pdr_override()
```

```python
mmm_data = pdr.get_data_yahoo(['MMM'], start='2022-01-01', end='2022-04-30')
hon_data = pdr.get_data_yahoo(['HON'], start='2022-01-01', end='2022-04-30')
kmb_data = pdr.get_data_yahoo(['KMB'], start='2022-01-01', end='2022-04-30')
apt_data = pdr.get_data_yahoo(['APT'], start='2022-01-01', end='2022-04-30')
```

```python
## 종목의 Open, High, Low, Close, Adj Close, Volume값 받아오기
print(mmm_data.to_string())
print("\n##############################\n")

print(hon_data.to_string())
print("\n##############################\n")

print(kmb_data.to_string())
print("\n##############################\n")

print(apt_data.to_string())
print("\n##############################\n")
```

```python
## 배당내역, 주식분할 이력, 애널리스트 평가 받아오기

# 0. 데이터를 받고자 하는 기업의 티커를 받아오기, 회사 요약 정보 불러오기(info)
mmm = yf.Ticker('MMM')
mmm_info = mmm.info
print(mmm_info)
print("##############################")

# 1. 배당 내역
print(mmm.dividends)
print("##############################")

# 2. 주식분할 이력
print(mmm.splits)
print("##############################")

# 3. 애널리스트 평가
print(mmm.recommendations)
print("##############################")

# 4. 회사 재무제표
print(mmm.financials)
print("##############################")

# 5. 주주정보
print(mmm.major_holders)
```
